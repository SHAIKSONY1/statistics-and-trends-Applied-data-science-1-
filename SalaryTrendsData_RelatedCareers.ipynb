```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# @title employment_type vs work_year

def employment_type(salary_df):
    figsize = (12, 1.2 * len(salary_df['employment_type'].unique()))
    plt.figure(figsize=figsize)
    sns.violinplot(data=salary_df, x='work_year', y='employment_type', inner='box', palette='Dark2')
    sns.despine(top=True, right=True, bottom=True, left=True)
    plt.show()

employment_type(salary_df)
```

```python
# @title experience_level

def experience_level(data, x_axis, y_axis):
    counts = data[x_axis].value_counts()
    ax = counts.plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))

    # Removing the right and top spines
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)

    # Adding count labels on top of each bar
    for i, count in enumerate(counts):
        plt.text(count, i, str(count), ha='left', va='center')

    plt.xlabel(y_axis)
    plt.ylabel(x_axis)
    plt.show()

experience_level(salary_df, 'experience_level', 'Count')
```

```python
# @title work_year vs salary

def plot_scatter_with_trend(data, x_column, y_column, hue_column=None, s=50, alpha=0.8):
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=data, x=x_column, y=y_column, hue=hue_column, s=s, alpha=alpha)
    sns.regplot(data=data, x=x_column, y=y_column, scatter=False)

    # Correlation coefficient
    correlation = data[x_column].corr(data[y_column])
    plt.text(0.95, 0.95, f'Correlation: {correlation:.2f}', ha='right', va='top', transform=plt.gca().transAxes, fontsize=10)

    # Spines
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)

    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.show()

plot_scatter_with_trend(salary_df, 'work_year', 'salary', hue_column='experience_level', s=50, alpha=0.8)
```

```python
# @title Correlation

def correlation_heatmap(data, figsize=(10, 8), cmap='coolwarm', annot=True, fmt=".2f", annot_kws={"size": 10}, title='Correlation Heatmap'):
    correlation_matrix = data.corr()

    plt.figure(figsize=figsize)
    sns.heatmap(correlation_matrix, annot=annot, cmap=cmap, fmt=fmt, annot_kws=annot_kws)

    plt.title(title)
    plt.show()

correlation_heatmap(salary_df)
```
